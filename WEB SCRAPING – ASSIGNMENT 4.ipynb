## 1. Scrape the details of most viewed videos on YouTube from Wikipedia.

```python
Url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

driver.get(Url)

#scrapping rank of the video
rank=[]
try:
    ranks=driver.find_elements(By.XPATH ,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[1]")
    for i in ranks:  
        rank.append(i.text)
        
except NoSuchElementException:
    rank.append("no detail available")
except StaleElementReferenceException:
    rank.append("no detail available")



#scrapping the name of the video
name=[]
try:
    names=driver.find_elements(By.XPATH ,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[2]")
    for i in names:
        name.append(i.text)
        
except NoSuchElementException:
    name.append("no detail available")
except StaleElementReferenceException:
    name.append("no detail available")
         

#scrapping the details of artist
artist=[]
try:
    artists=driver.find_elements(By.XPATH ,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]")
    for i in artists:
        artist.append(i.text)
        
except NoSuchElementException:
    artist.append("no detail available")
except StaleElementReferenceException:
    artist.append("no detail available")
   


#scrapping the details of upload date
upload_date=[]
try:
    upload_dates_list=driver.find_elements(By.XPATH ,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[5]")
    for i in upload_dates_list:
        upload_date.append(i.text)
        
except NoSuchElementException:
    upload_date.append("no detail available")
except StaleElementReferenceException:
    upload_date.append("no detail available")
    

#scrapping the details of views
views=[]
try:
    view=driver.find_elements(By.XPATH ,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[4]")
    for i in view:
        views.append(i.text)
        
except NoSuchElementException:
    views.append("no detail available")
except StaleElementReferenceException:
    views.append("no detail available")
        

#preparing dataframe
videos=pd.DataFrame({"Rank":rank,"Name":name,"Artist":artist,"Upload Date":upload_date,"Views":views})
#videos.to_csv("Top_30_Videos",index=True)
print(videos)

```

## 2. Scrape the details team India’s international fixtures from bcci.tv.

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.maximize_window()

#getting the specified url
url = "https://www.bcci.tv/"
driver.get(url)
time.sleep(5)

#Getting url for international fixture and clicking on it
Options_btn = driver.find_element(By.XPATH , '//div[@class="collapse navbar-collapse"]/ul/li[2]').click()
time.sleep(3)

try:
    driver.find_element(By.XPATH , '//button[@class="cookie__accept btn btn-primary"]').click()
except NoSuchElementException:
    pass

more_fixtures = driver.find_element(By.XPATH , '//button[@class="match-btn btn-red d-flex align-items-center justify-content-center mx-auto mt-3"]').click()
#Infix_btn = driver.find_element(By.XPATH ,"//a[@class='navigation__link navigation__link--in-drop-down']").click()
time.sleep(3)

#Creating empty list
Match_title = []
Series = []
Place = []
Date = []
Time = []

#Scraping data of match title
for i in driver.find_elements(By.XPATH ,'//span[@class="matchOrderText ng-binding ng-scope"]'):
    i = i.text[:-2]
    Match_title.append(i)
    
#Scraping data of Place
for i in driver.find_elements(By.XPATH ,'//div[@class="fix-place ng-binding ng-scope"]'):
    i = i.text
    x = i.split('-')
    Place.append(x[1])
 
#scraping series data
series = driver.find_elements(By.XPATH ,'//span[@class="ng-binding"]')

for i in series:
    try:
        Series.append(i.text)
    except NoSuchElementException:
        Series.append('-')


#Scraping data of dates

dates = driver.find_elements(By.XPATH ,'//div[@class="match-card-left match-schedule"]')
for i in dates:
    try:
        Date.append(i.text)
    except NoSuchElementException:
        Date.append('-')

#Scraping time data
times = driver.find_elements(By.XPATH ,'//div[@class="match-card-right match-schedule "]')
for i in times:
    try:
        Time.append(i.text)
    except NoSuchElementException:
        Time.append('-')
        
#Creating dataframe
BCCI=pd.DataFrame({})
BCCI['Match Title'] = Match_title
BCCI['Series'] = Series
BCCI['Place'] = Place
BCCI['Date'] = Date
BCCI['Time'] = Time

#Printing data frame
BCCI
```

## 4. Scrape the details of State-wise GDP of India from statisticstime.com.

#### I could not find GSDP 17-18. hence I scraped GSDP 19-20, assuming the question paper is slightly outdated

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.maximize_window()

#getting the specified url
url = "http://statisticstimes.com/"
driver.get(url)
time.sleep(3)

#Clicking on economy section
economy = driver.find_element(By.XPATH ,"//div[@class='navbar']//div[2]//button").click()

urls = driver.find_element(By.XPATH ,"//div[@class='dropdown-content']//a[3]")
ineco_page = urls.get_attribute("href")
#Going to indian economy page
driver.get(ineco_page)  
time.sleep(5)

#Creating empty list
Rank = []
State = []
GSDP_1 = []
GSDP_2 = []
Share = []
GDP = []

#get the url of page containing GDP of indian states
sta_GDP = driver.find_element(By.XPATH ,"//ul[@style='list-style-type:none;margin-left:20px;']/li/a")
GDP_url = sta_GDP.get_attribute("href")

driver.get(GDP_url)
time.sleep(6)


#Scraping data of rank
for i in driver.find_elements(By.XPATH ,"//div[@id='table_id_wrapper']//tbody//tr//td[1]"):
    Rank.append(i.text)

    
#Scraping data of state name
for i in driver.find_elements(By.XPATH ,"//div[@id='table_id_wrapper']//tbody//tr//td[2]"):
    State.append(i.text)

    
# Scraping data of GSDP at current price (19-20)
for i in driver.find_elements(By.XPATH ,"//div[@id='table_id_wrapper']//tbody//tr//td[3]"):
    GSDP_1.append(i.text)

    
#Scraping data of GSDP at current price (18-19)
for i in driver.find_elements(By.XPATH ,"//div[@id='table_id_wrapper']//tbody//tr//td[4]"):
    GSDP_2.append(i.text)

    
#Scraping data of Share(18-19)
for i in driver.find_elements(By.XPATH ,"//div[@id='table_id_wrapper']//tbody//tr//td[5]"):
    Share.append(i.text)

    
#Scraping data of GDP($ billion)
for i in driver.find_elements(By.XPATH ,"//div[@id='table_id_wrapper']//tbody//tr//td[6]"):
    GDP.append(i.text)
    
#DATA FRAMEING
StatisticsTime=pd.DataFrame({})
StatisticsTime['Rank'] = Rank
StatisticsTime['State'] = State
StatisticsTime['GSDP at current price (19-20)'] = GSDP_1
StatisticsTime['GSDP at current price (18-19)'] = GSDP_2
StatisticsTime['Share(18-19)'] = Share
StatisticsTime['GDP($ billion)'] = GDP

#Printing the data frame
StatisticsTime
```

## 5. Scrape the details of trending repositories on Github.com.

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.maximize_window()

#getting the specified url
url = "https://github.com/"
driver.get(url)
time.sleep(3)

#Getting explore button and clicking on it
explore = driver.find_element(By.XPATH ,'//nav[@class="mt-0 px-3 px-lg-0 mb-3 mb-lg-0"]/ul/li[3]').click()

#Selecting trending option
trend_url = driver.find_element(By.XPATH ,'//nav[@class="mt-0 px-3 px-lg-0 mb-3 mb-lg-0"]/ul/li[3]/div/ul[3]/li[3]/a')
urls = trend_url.get_attribute("href")
driver.get(urls)

#Creating empty list
repo_urls = []
rep_title = []
Description = []
Contributors = []
Language = []
lang = []

#Fetching urls for each repository
repository = driver.find_elements(By.XPATH ,"//h1[@class = 'h3 lh-condensed']//a")
for i in repository:
    repo_urls.append(i.get_attribute("href"))

    
#Scraping data of repository title
title = driver.find_elements(By.XPATH ,"//h1[@class = 'h3 lh-condensed']")
for i in title:
    rep_title.append(i.text)

    
#Scraping data from every repository page
for i in repo_urls:
    driver.get(i)
    time.sleep(3)
    
    
    #Scraping data of decription
    try:
        desc = driver.find_element(By.XPATH ,'//article[@class="markdown-body entry-content container-lg"]/p[2]')
        Description.append(desc.text)
    except NoSuchElementException:
        Description.append('-')
        
        
    #Scraping data of contributors
    try:
        cont_tags = driver.find_element(By.XPATH ,"//*[contains(text(),'    Contributors ')]")
        Contributors.append(cont_tags.text.replace('Contributors',''))
    except NoSuchElementException:
        Contributors.append('-')
        
        
    #fetching Languages used
    try:
        for i in driver.find_elements(By.XPATH ,'//div[@class="Layout-sidebar"]/div/div[6]/div/ul/li/a/span[1]'):
            lang.append(i.text)
        Language.append(lang)
    
    except NoSuchElementException:
        Language.append('-')
        
#DATA FRAMEING
Github=pd.DataFrame({})
Github['Repository title'] = rep_title
Github['Repository description'] = Description
Github['Contributors count'] = Contributors
Github['Language used'] = Language

#Printing data frame
Github
```

## 8. Scrape the details of Highest selling novels.

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.maximize_window()

#getting the specified url
url = "https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare/"
driver.get(url)
time.sleep(3)

#Creating empty list
Book_name = []
Author = []
Volumes_sold = []
Publisher = []
Genre = []

#scraping data of book names
for i in driver.find_elements(By.XPATH ,"//tbody//tr/td[2]"):
    Book_name.append(i.text)

    
#Scraping data of author names
for i in driver.find_elements(By.XPATH ,"//tbody/tr/td[3]"):
    try:
        if i.text == '0' : raise NoSuchElementException
        Author.append(i.text)
    except NoSuchElementException:
        Author.append('-')
time.sleep(1)


#Scraping data of volumes sold
for i in driver.find_elements(By.XPATH ,"//tbody/tr/td[4]"):
    Volumes_sold.append(i.text)

    
#Scraping data of publisher names
for i in driver.find_elements(By.XPATH ,"//tbody/tr/td[5]"):
    Publisher.append(i.text)

    
#Scraping data of genre
for i in driver.find_elements(By.XPATH ,"//tbody/tr/td[6]"):
    Genre.append(i.text)
    
#DATA FRAMEING
Novel=pd.DataFrame({})
Novel['Book Name'] = Book_name
Novel['Author'] = Author
Novel['Volume sold'] = Volumes_sold
Novel['Publisher'] = Publisher
Novel['Genre'] = Genre

#Printing data frame
Novel
```

## 9. Scrape the details most watched tv series of all time from imdb.com.

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.maximize_window()

#getting the specified url
url = "https://www.imdb.com/list/ls095964455/"
driver.get(url)
time.sleep(3)

#Creating empty lists.
Name = []
Year_span = []
Genre = []
Run_time = []
Ratings = []
Votes = []

#Scraping data of Names
for i in driver.find_elements(By.XPATH ,"//h3[@class='lister-item-header']/a"):
    Name.append(i.text)

    
#Scraping data of Year span
for i in driver.find_elements(By.XPATH ,"//span[@class='lister-item-year text-muted unbold']"):
    Year_span.append(i.text)

    
#Scraping data of genre
for i in driver.find_elements(By.XPATH ,"//span[@class='genre']"):
    Genre.append(i.text)

    
#Scraping data of Run time
for i in driver.find_elements(By.XPATH ,"//span[@class='runtime']"):
    Run_time.append(i.text)

    
#Scraping data of Ratings
for i in driver.find_elements(By.XPATH ,"//div[@class='ipl-rating-star small']//span[2]"):
    Ratings.append(i.text)

    
#Scraping data of votes
for i in driver.find_elements(By.XPATH ,"//div[@class='lister-item-content']//p[4]/span[2]"):
    Votes.append(i.text)
    
#DATA FRAMEING
IMDB=pd.DataFrame({})
IMDB['Name'] = Name
IMDB['Year Span'] = Year_span
IMDB['Genre'] = Genre
IMDB['Run Time'] = Run_time
IMDB['Ratings'] = Ratings
IMDB['Votes'] = Votes

#Printing data frame
IMDB
```

## 10. Details of Datasets from UCI machine learning repositories.

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.maximize_window()

#getting the specified url
url = "https://archive.ics.uci.edu/"
driver.get(url)
time.sleep(3)

#Finding view all dataset button
view_dataset = driver.find_element(By.XPATH ,"//tbody[1]//tr/td[2]/span[2]/a")    
page_url = view_dataset.get_attribute("href")
driver.get(page_url)
time.sleep(5)

#Getting page urls containing list of all datasets
all_lst = driver.find_element(By.XPATH ,"/html/body/table[2]/tbody/tr/td[2]/table[1]/tbody/tr/td[2]/p/a")  
lst_url = all_lst.get_attribute("href")           
driver.get(lst_url)
time.sleep(5)

#fetching url for each dataset
data_url = driver.find_elements(By.XPATH ,"//p[@class='normal']//b/a")    

urls = []     
for i in data_url:
    urls.append(i.get_attribute("href"))
    
#Creating empty list
Data_name = []
Data_type = []
Task = []
Attr_type = []
Instances = []
n_attributes = []
Year = []

#Scraping requiredinfo from urls
for i in urls:
    driver.get(i)
    time.sleep(3)
    
    
    #Scraping Dataset name
    try: 
        ds_name = driver.find_element(By.XPATH ,"//span[@class='heading']")
        Data_name.append(ds_name.text)
    except NoSuchElementException:
        Data_name.append('-')
    time.sleep(1)
    
    
    #Scraping data type
    try:
        dtype = driver.find_element(By.XPATH ,"//table[@border='1']//tbody/tr/td[2]")
        if dtype.text == "N/A": raise NoSuchElementException
        Data_type.append(dtype.text)
    except NoSuchElementException:
        Data_type.append('-')
    time.sleep(1)
    
    
    #Scraping Attribute type
    try:
        atype = driver.find_element(By.XPATH ,"//table[@border='1']//tbody/tr[2]/td[2]")
        if atype.text == "N/A": raise NoSuchElementException
        Attr_type.append(atype.text)
    except NoSuchElementException:
        Attr_type.append('-')
    time.sleep(1)
    
    
    #scraping Task
    try:
        task = driver.find_element(By.XPATH ,"//table[@border='1']//tbody/tr[3]/td[2]")
        if task.text == "N/A": raise NoSuchElementException
        Task.append(task.text)
    except NoSuchElementException:
        Task.append('-')
    time.sleep(1)
    
    
    #Scraping No of instances
    try:
        inst = driver.find_element(By.XPATH ,"//table[@border='1']//tbody/tr/td[4]")
        if inst.text == "N/A": raise NoSuchElementException
        Instances.append(inst.text)
    except NoSuchElementException:
        Instances.append('-')
    time.sleep(1)
    
    
    #Scraping  No of attribute
    try:
        attr = driver.find_element(By.XPATH ,"//table[@border='1']//tbody/tr[2]/td[4]")
        if attr.text == "N/A": raise NoSuchElementException
        n_attributes.append(attr.text)
    except NoSuchElementException:
        n_attributes.append('-')
    time.sleep(1)
    
    
    #scraping year
    try:
        year = driver.find_element(By.XPATH ,"//table[@border='1']//tbody/tr[2]/td[6]")
        if year.text == "N/A": raise NoSuchElementException
        Year.append(year.text[:4])
    except NoSuchElementException:
        Year.append('-')
    time.sleep(1)
    
#DATA FRAMEING
UCI=pd.DataFrame({})
UCI['Data Name'] = Data_name
UCI['Data Type'] = Data_type
UCI['Task'] = Task
UCI['Attribute type'] = Attr_type
UCI['No of instance'] = Instances
UCI['No of attributes'] = n_attributes
UCI['Year'] = Year

#Printing dataframe
UCI
```
