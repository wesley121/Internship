### Q1: Write a python program which searches all the product under a particular product vertical from www.amazon.in. The product verticals to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search for guitars.

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
url = "https://www.amazon.in/"
driver.get(url)

driver.maximize_window()

time.sleep(3)

search_product = driver.find_element(By.XPATH , '//input[@id="twotabsearchtextbox"]')
search_product.send_keys("Guitar")

search_button = driver.find_element(By.XPATH , '//input[@ id="nav-search-submit-button"]')
search_button.click()
time.sleep(3)

Brand_name = []
Product_name = []
Price = []
Exchange = []
Estimated_delivery = []
Availability = []
Product_link = []
    
for i in range(3):
    
    product_title_tags = driver.find_elements(By.XPATH , '//span[@class="a-size-base-plus a-color-base a-text-normal"]')
    no_of_products = len(product_title_tags)

    for i in range(no_of_products + 1):
        if i == 10:
            continue

        product_title_tags[i].click()
        time.sleep(8)

        parent = driver.window_handles[0]
        child = driver.window_handles[1]

        driver.switch_to.window(child)
       

        #Getting Brand name
        try:
            brand_tag_1 = driver.find_elements(By.XPATH, '//tr[@class="a-spacing-none a-spacing-top-small po-brand"]')
            brand_tag_2 = driver.find_elements(By.XPATH, '//tr[@class="a-spacing-small po-brand"]')

            if len(brand_tag_1) == 1:
                brand_name_tag = driver.find_element(By.XPATH, '//tr[@class="a-spacing-none a-spacing-top-small po-brand"]/td[2]/span')
                brand_name = brand_name_tag.text


            elif len(brand_tag_2) == 1:
                brand_name_tag = driver.find_element(By.XPATH, '//tr[@class="a-spacing-small po-brand"]/td[2]/span')
                brand_name = brand_name_tag.text
            Brand_name.append(brand_name)
        except NoSuchElementException:
            Brand_name.append('-') 


        #getting name of the product
        try:
            name_of_the_product = driver.find_element(By.XPATH , '//span[@id="productTitle"]').text
            Product_name.append(name_of_the_product)
        except NoSuchElementException:
            Exchange.append('-') 
        
        #getting price
        try:
            price = driver.find_element(By.XPATH , '//div[@id="apex_offerDisplay_desktop"]/div/div/span/span[2]').text
            Price.append(price)
        except NoSuchElementException:
            Exchange.append('-')
        
        #getting Exchange policy
        try:
            exchange = driver.find_element(By.XPATH , '//div[@class="a-row icon-farm-wrapper"]/div[2]/span/div[2]/a').text
            Exchange.append(exchange)
        except NoSuchElementException:
            Exchange.append('-')
        
        #Getting estimated delivery
        try:
            expected_delivery = driver.find_element(By.XPATH , '//div[@ class="a-spacing-base"][2]').text
            Estimated_delivery.append(expected_delivery)
        except NoSuchElementException:
            Exchange.append('-')
        
        #Getting availability
        try:
            availability = driver.find_element(By.XPATH , '//div[@id="availability"]').text
            Availability.append(availability)
        except NoSuchElementException:
            Exchange.append('-')
        
        #Getting product link
        try:
            share_button = driver.find_element(By.XPATH , '//i[@title="Share"]')
            share_button.click()
            time.sleep(2)
            copy_link = driver.find_element(By.XPATH , '//span[@class="copy-label"]')
            copy_link.click()
            time.sleep(2)

            root = tk.Tk()
            product_link = root.clipboard_get()
            Product_link.append(product_link)
        except NoSuchElementException:
            Exchange.append('-')



        driver.close()
        driver.switch_to.window(parent)
        
    
    next_button = driver.find_element(By.XPATH , '//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
    next_button.click()
    time.sleep(5)
    
driver.close()



```

## Question 5 : Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps.

```python
import re

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# getting mentioned url and opening google maps web page
url = 'https://www.google.co.in/maps'
driver.get(url)
driver.maximize_window()
time.sleep(2)

# entering the city name in search bar
City = input('Enter City name that has to be searched : ')
search_bar = driver.find_element(By.XPATH,'//input[@class="tactile-searchbox-input"]')
search_bar.click()
time.sleep(2)

#sending keys to find cities
search_bar.send_keys(City)

#checking for webelement and clicking on search button
search_btn = driver.find_element(By.XPATH, '//button[@id="searchbox-searchbutton"]')
search_btn.click()
time.sleep(2)

try:
    url_str = driver.current_url
    print("URL Extracted: ", url_str)
    latitude_longitude = re.findall(r'@(.*)data',url_str)
    if len(latitude_longitude):
        lat_lng_list = latitude_longitude[0].split(",")
        if len(lat_lng_list)>=2:
            latitude = lat_lng_list[0]
            longitude = lat_lng_list[1]
        print("Latitude = {}, Longitude = {}".format(latitude, longitude))
except Exception as e:
        print("Error: ", str(e))
 
```
