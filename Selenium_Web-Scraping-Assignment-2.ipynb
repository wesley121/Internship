## Question 1 : Scraping for  job-title, job-location, company_name, experience_required

### experience_required isnt an option on the job listings on jobstreet. Hence I am scraping for "posted time" instead

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
url = "https://www.jobstreet.com.my/"
driver.get(url)
time.sleep(5)


# finding element for job search bar

search_job = driver.find_element(By.XPATH, "//input[@ id = 'searchKeywordsField']")  
search_job.send_keys('Data Analyst')

# finding element for job location bar
search_loc = driver.find_element(By.XPATH, "//input[@ placeholder='Area, city or town']")  
search_loc.send_keys("Kuala lumpur")

# clicking  using xpath function
search_btn = driver.find_element(By.XPATH, "//button[@ data-automation='searchSubmitButton']")  
search_btn.click()

#so let's extract all the tags having the job titles

title_tags = driver.find_elements(By.XPATH , "//div[@ class = 'sx2jih0 l3gun70 l3gun74 l3gun72']/span")
time.sleep(5)

# extract the text of the job title  from the tags
job_titles=[]
for i in title_tags:
    if i.text is None:
        job_titles.append('Not')
    else:
        job_titles.append(i.text)
   

# lets extract all the tags having company names
company_tags = driver.find_elements(By.XPATH , "//span[@ class = 'sx2jih0 zcydq84u _18qlyvc0 _18qlyvc1x _18qlyvc1 _18qlyvc7']/span/a")

# Now we will extract the text from the tags by looping over these tags
companies_names=[]

for i in company_tags:
    companies_names.append(i.text)


# so lets extract  all the tags having the job location

location_tags = driver.find_elements(By.XPATH ,"//span[@ class = 'sx2jih0 zcydq84u _18qlyvc0 _18qlyvc1x _18qlyvc3 _18qlyvc7']/span/a")

# now we will extract the text  from these tags only by one by looping over these tags
location_list=[]
for i in location_tags:
    location_list.append(i.text)


#extracting when the job posting was posted
posted_tags = driver.find_elements(By.XPATH ,'//time[@ class="sx2jih0 zcydq84u"]/span')

posted_list=[]
for i in posted_tags:
    posted_list.append(i.text)

jobs=pd.DataFrame({})
jobs['title']=job_titles[:10]
jobs['company']=companies_names[:10]
jobs['location']=location_list[:10]
jobs['Posted']=posted_list[:10]

driver.close()

jobs

```

## 3. using filters to scrape jobs in Delhi/ NCR, salary range : 3-6 Lakhs

### Since the filters that this question wanted to test me on, were not available on jobstreet, I developed this solution for Naukri. unfortunately, I could not test it because im in malaysia and the site is blocking me for being a bot. Please dont mind if encountered with some errors.

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

url = "https://www.naukri.com"
driver.get(url)
time.sleep(5)


# finding element for job search bar

search_job = driver.find_element(By.XPATH, "//input[@class='sugInp']")  
search_job.send_keys('Data Scientist')

# finding element for job location bar
# search_loc = driver.find_element(By.XPATH, "//input[@ placeholder='Area, city or town']")  
# search_loc.send_keys("Kuala lumpur")

# clicking  using xpath function
search_btn = driver.find_element(By.XPATH, "//button[@class='btn']")  
search_btn.click()
time.sleep(5)
#so let's extract all the tags having the job titles


filter_loc = driver.find_element(By.XPATH, '//span[@title="Delhi / NCR"]')
filter_loc.click()

filter_salary = driver.find_element(By.XPATH, '//span[@title="3-6 Lakhs"]')
filter_salary.click()



title_tags = driver.find_elements(By.XPATH , '//a[@ class="title fw500 ellipsis"]')

# extract the text of the job title  from the tags
job_titles=[]
for i in title_tags:
    if i.text is None:
        job_titles.append('Not')
    else:
        job_titles.append(i.text)
   

# lets extract all the tags having company names
company_tags = driver.find_elements(By.XPATH , '//a[@class="subTitle ellipsis fleft"]')

# Now we will extract the text from the tags by looping over these tags
companies_names=[]

for i in company_tags:
    companies_names.append(i.text)


# so lets extract  all the tags having the job location

location_tags = driver.find_elements(By.XPATH ,'//li[@class="fleft grey-text br2 placeHolderLi location"]')

# now we will extract the text  from these tags only by one by looping over these tags
location_list=[]
for i in location_tags:
    location_list.append(i.text)


#extracting when the job posting was posted
exp_tags = driver.find_elements(By.XPATH ,'//li[@class="fleft grey-text br2 placeHolderLi experience"]/i')

exp_list=[]
for i in exp_tags:
    exp_list.append(i.text)

jobs=pd.DataFrame({})
jobs['title']=job_titles[:10]
jobs['company']=companies_names[:10]
jobs['location']=location_list[:10]
jobs['Experience needed']=exp_list[:10]

driver.close()

jobs
```

## 5. Rating, review summary and full review of 100 iphone 11 reviews on flipkart

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

url = "https://www.flipkart.com/"
driver.get(url)
time.sleep(10)

esc_login = driver.find_element(By.XPATH , '//button[@class="_2KpZ6l _2doB4z"]')
esc_login.click()

search_product = driver.find_element(By.XPATH, '//input[@class="_3704LK"]')  
search_product.send_keys('iphone 11')

search_btn = driver.find_element(By.XPATH, '//button[@class="L0Z3Pu"]')  
search_btn.click()
time.sleep(5)

product = driver.find_element(By.XPATH, '//div[@class="col col-7-12"]')
product.click()
time.sleep(5)

p = driver.current_window_handle
parent = driver.window_handles[0]
child = driver.window_handles[1]

driver.switch_to.window(child)


show_all_reviews = driver.find_element(By.XPATH, '//div[@class="_3UAT2v _16PBlm"]/span')
show_all_reviews.click()
time.sleep(5)

Rating=[]
Review_summary=[]
Full_Review=[]


for i in range(10):
     
    
    rating_tags = driver.find_elements(By.XPATH, '//div[@class="_3LWZlK _1BLPMq"]')
    review_summary_tags = driver.find_elements(By.XPATH, '//p[@class="_2-N8zT"]')
    full_review_tags = driver.find_elements(By.XPATH, '//div[@class="t-ZTKy"]/div/div')


    for rating in rating_tags:
        Rating.append(rating.text)
        
    for review in review_summary_tags:
        Review_summary.append(review.text)
    
    for review in full_review_tags:
        Full_Review.append(review.text)
    
    time.sleep(5)
    
    next_button = driver.find_element(By.XPATH, '//a[@class="_1LKTO3"]')
    next_button.click()
    
    time.sleep(5)
    
reviews = pd.DataFrame({})
reviews['Rating']=Rating
reviews['Review_summary'] = Review_summary
reviews['Full_Review'] = Full_Review

driver.close()

reviews

```

## 7. Filtering for black shoes on Myntra

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

url="https://www.myntra.com/shoes"
driver.get(url)
time.sleep(10)

#Applying the black colur filter
filter_button2 = driver.find_element(By.XPATH,'//li[@class="colour-listItem"][1]')
filter_button2.click()
time.sleep(5)  

# appplying the price filter
filter_button1=driver.find_element(By.XPATH,'//ul[@class="price-list"]/li[2]/label')
filter_button1.click()
time.sleep(5)


Brand=[]
Price=[]
Description=[]

for i in range(2):
    brand=driver.find_elements(By.XPATH,'//h3[@class="product-brand"]')
    description=driver.find_elements(By.XPATH,'//h4[@class="product-product"]')
    price =driver.find_elements(By.XPATH,'//div[@class="product-price"]')
 
    for j  in brand:
        Brand.append(j.text)
    
    for k in description:
        Description.append(k.text)
    
    for l in price:
        temp = l.text.split('Rs.' , 2)
        if len(temp) == 3:
            l = temp[1]
        else: l = temp[1]
        Price.append(l)


myntra_shoes = pd.DataFrame()
myntra_shoes["Brand"] = Brand
myntra_shoes['Description'] = Description
myntra_shoes['Price in INR'] = Price

driver.close()

myntra_shoes
```

## 9. Ambitionbox.com (Data scientist - Noida)

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))


url= "https://www.ambitionbox.com/"
driver.get(url)
time.sleep(10)

jobs = driver.find_element(By.XPATH,'//a[@title="Jobs"]')
jobs.click()
time.sleep(5)

search = driver.find_element(By.NAME,'ab_jobsSearch')
search.send_keys('Data Scientist')

search_btn = driver.find_element(By.XPATH,'//div[@class="fixed-typeahead-wrapper"]/div/div/button')
search_btn.click()
time.sleep(5)

location = driver.find_element(By.XPATH, '//div[@title="Location"]')
location.click()

noida = driver.find_element(By.XPATH, '//label[@for="location_Noida"]')
noida.click()
time.sleep(5)

company_name_tags = driver.find_elements(By.XPATH, '//a[@class="title noclick"]')
ratings_tags = driver.find_elements(By.XPATH, '//span[@class="body-small"]')
posted_tags = driver.find_elements(By.XPATH, '//div[@class="other-info"]/span[1]')


Company_name = []
Rating = []
Posted_on = []

for tag in company_name_tags:
    Company_name.append(tag.text)
    
for tag in ratings_tags:
    Rating.append(tag.text)
    
for tag in posted_tags:
    Posted_on.append(tag.text)
    

jobs = pd.DataFrame()

jobs["Company_name"] = Company_name
jobs["Rating"] = Rating
jobs["Posted_on"] = Posted_on

driver.close()

jobs
```
